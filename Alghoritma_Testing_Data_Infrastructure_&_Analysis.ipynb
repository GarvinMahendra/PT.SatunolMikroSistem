**1.** Anda memiliki dua buah array, yaitu DATA dan QUERIES, tentukan berapa kali setiap kata dalam QUERIES muncul dalam array DATA. Buatlah sebuah fungsi atau program untuk menghasilkan output yang berisi jumlah kemunculan setiap kata dalam QUERIES dalam DATA. Contoh:

  DATA = ['aple', 'banana', 'cherry', 'banana', 'apple']
  
  QUERIES = ['apple', 'banana', 'grape']

  -  Fungsi atau program Anda akan menghasilkan output berupa list, seperti ini:
    
    [2, 2, 0]

  - Penjelasan:
    - Kata 'apple' muncul 2 kali dalam DATA.
    - Kata 'banana' juga muncul 2 kali dalam DATA.
    - Kata 'grape' tidak muncul sama sekali dalam DATA.

  **JAWABAN**

```python
def hitung_kemunculan_kata(DATA, QUERIES):
    hasil = []

    for kata in QUERIES:
        kemunculan = DATA.count(kata)
        hasil.append(kemunculan)

    return hasil

# Contoh penggunaan
DATA = ['apple', 'banana', 'cherry', 'banana', 'apple']
QUERIES = ['apple', 'banana', 'grape']
output = hitung_kemunculan_kata(DATA, QUERIES)
print(output)
```

**2.** Silahkan cari hasil dari pengurangan dari jumlah diagonal sebuah matrik NxN Contoh:

  Matrix = [[1, 2, 0], [4, 5, 6], [7, 8, 9]]

  Diagonal pertama = 1 + 5 + 9 = 15

  Diagonal kedua = 0 + 5 + 7 = 12

  Maka hasilnya adalah 15 - 12 = 3

  **JAWABAN**

```python
def selisih_diagonal(matriks):
    ukuran = len(matriks)
    diagonal_pertama = 0
    diagonal_kedua = 0

    for i in range(ukuran):
        diagonal_pertama += matriks[i][i]
        diagonal_kedua += matriks[i][ukuran - 1 - i]

    hasil = diagonal_pertama - diagonal_kedua
    return hasil

# Contoh penggunaan
matriks = [[1, 2, 0], [4, 5, 6], [7, 8, 9]]
hasil = selisih_diagonal(matriks)
print("Hasil selisih diagonal:", hasil)
```

```python
import folium

# Buat peta dengan koordinat pusat Jakarta
jakarta_map = folium.Map(location=[-6.2088, 106.8456], zoom_start=10)

# Koordinat segitiga
coordinates = [
    [-6.1, 106.7],
    [-6.1, 106.9],
    [-6.3, 106.8],
    [-6.1, 106.7]  # Titik penutup segitiga (kembali ke titik awal)
]

# Warna tiap garis segitiga
colors = ['red', 'blue', 'green']

# Buat segitiga dengan tiga garis berwarna berbeda
for i in range(3):
    folium.PolyLine(
        locations=coordinates[i:i+2],  # Ambil dua titik sekaligus
        color=colors[i],
        weight=3  # Lebar garis
    ).add_to(jakarta_map)

# Tampilkan peta
jakarta_map.save("colored_triangles.html")
```
